```python
dimension = 1000000
y = np.random.randn(dimension)
print y[:10]
z = np.random.randn(dimension)
print z[:10]
```

```python
# inner product in a pythonic way

t3 = time.time()
innProd = 0
for i in xrange(dimension):
    innProd += y[i]*z[i]
    
t4 = time.time()
    
print innProd
print 'time taken =', t4-t3, 'secs'
```

```python
# inner product in numpy

t5 = time.time()
innProd2 = np.dot(y,z)
t6 = time.time()

print innProd2
print 'time taken =', t6-t5, 'secs'
```

```python
# generating random variables of a particular distribution

x = np.random.rand(2,500)
print x
print x[0,:] # can be thought of as the x-coordinate
print x[1,:] # can be thought of as the y-coordinate
```

```python
import matplotlib.pyplot as plt

plt.plot(x[0,:], x[1,:], 'rx')
plt.axis([0,1,0,1])
plt.show()
```

```python
xn = 0.25*np.random.randn(2,500)
print xn
print xn[0,:] # can be thought of as the x-coordinate
print xn[1,:] # can be thought of as the y-coordinate
```

```python
plt.plot(xn[0,:], xn[1,:], 'bx')
plt.axis([-1,1,-1,1])
plt.show()
```

```python
# histogram for random variables

numOfSamples = 100000
z = 0.25*np.random.randn(numOfSamples)
hist, bins = np.histogram(z, bins=150)
print hist
print bins
```

```python
bins[:-1] + 0.5 * (bins[1]-bins[0])
```

```python
np.diff(bins)
```

```python
x = bins[:-1] + 0.5 * (bins[1]-bins[0])
y = hist / float(numOfSamples)
print y
```

```python
plt.plot(x,y)
plt.axis([-1,1,0,np.max(y)])
```

```python
# similar for other distributions

a = 4
b = 2
z = np.random.beta(a,b, numOfSamples)
hist, bins = np.histogram(z, bins=150)
x = bins[:-1] + 0.5 * (bins[1]-bins[0])
y = hist / float(numOfSamples)
plt.plot(x,y)
plt.axis([0,1,0,np.max(y)])
```
